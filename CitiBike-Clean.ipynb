```python
# dependencies
import pandas as pd
import csv
import os
```

```python
# read in CSV files
jan = os.path.join(os.getcwd(), "Resources", "JC-202001-citibike-tripdata.csv")
feb = os.path.join(os.getcwd(), "Resources", "JC-202002-citibike-tripdata.csv")
mar = os.path.join(os.getcwd(), "Resources", "JC-202003-citibike-tripdata.csv")
apr = os.path.join(os.getcwd(), "Resources", "JC-202004-citibike-tripdata.csv")
may = os.path.join(os.getcwd(), "Resources", "JC-202005-citibike-tripdata.csv")
jun = os.path.join(os.getcwd(), "Resources", "JC-202006-citibike-tripdata.csv")
jul = os.path.join(os.getcwd(), "Resources", "JC-202007-citibike-tripdata.csv")
aug = os.path.join(os.getcwd(), "Resources", "JC-202008-citibike-tripdata.csv")
sep = os.path.join(os.getcwd(), "Resources", "JC-202009-citibike-tripdata.csv")
oct = os.path.join(os.getcwd(), "Resources", "JC-202010-citibike-tripdata.csv")
nov = os.path.join(os.getcwd(), "Resources", "JC-202011-citibike-tripdata.csv")
dec = os.path.join(os.getcwd(), "Resources", "JC-202012-citibike-tripdata.csv")
```

```python
# create dfs from csvs
jan_df = pd.read_csv(jan)
feb_df=pd.read_csv(feb)
mar_df=pd.read_csv(mar)
apr_df=pd.read_csv(apr)
may_df=pd.read_csv(may)
jun_df=pd.read_csv(jun)
jul_df=pd.read_csv(jul)
aug_df=pd.read_csv(aug)
sep_df=pd.read_csv(sep)
oct_df=pd.read_csv(oct)
nov_df=pd.read_csv(nov)
dec_df=pd.read_csv(dec)
```

```python
# join dfs
citibike_2020_df = pd.concat([jan_df, feb_df, mar_df, apr_df, may_df, jun_df, jul_df, aug_df, sep_df, oct_df, nov_df, dec_df])
citibike_2020_df.head()
```

```python
# export new data to CSV
citibike_2020_df.to_csv("2020-citibike-data.csv")
```
